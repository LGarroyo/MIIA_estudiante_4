* *Milla extra*: escribe una función que sea capaz de hacer lo anterior para un archivo y úsala en un ciclo para automatizar la lectura de los diferentes archivos.

```python
# código
#from google.colab import drive
#drive.mount("/content/gdrive") ##Esto porque trabajé en google colab

##Creamos la función
def lectura_archivos (direccion, nombre_archivo):
  variable=[]
  with open(direccion+nombre_archivo, "r") as mfh: 
    lista_mfh = mfh.read()
    # Separamos su contenido por líneas
    lista_mfh = lista_mfh.splitlines()
    for i in lista_mfh:
      variable.append(i)
  return variable

###############  
##Especificamos la direccion de los archivos
dir="/content/gdrive/My Drive/Colab Notebooks/Archivos/"
###Especificamos el nombre de todos los arcivos
arc=["edad.txt", "escolaridad.txt", "estado_civil.txt", "estrato.txt", "genero.txt", "promedio.txt", "region.txt"]
##Creamos diccionario que contendra las variables, y cuya clave sera el nombre de la variable
datos={}
for i in arc:
  datos[i[:-4]]=lectura_archivos(dir, i)

```
